```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("company.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER
)
""")

conn.commit()
print("Table created successfully!")
```

```python
cursor.execute("INSERT INTO employees VALUES (1,'John','HR',50000)")
cursor.execute("INSERT INTO employees VALUES (2,'Alice','IT',70000)")
cursor.execute("INSERT INTO employees VALUES (3,'Bob','Finance',60000)")
cursor.execute("INSERT INTO employees VALUES (4,'David','IT',80000)")

conn.commit()
print("Data inserted successfully!")
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees WHERE salary > 60000",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees ORDER BY salary DESC",
    conn
)

df
```

```python
df = pd.read_sql_query(
    """
    SELECT department,
           AVG(salary) AS average_salary
    FROM employees
    GROUP BY department
    """,
    conn
)

df
```

```python
df = pd.read_sql_query(
    """
    SELECT department,
           AVG(salary) AS average_salary
    FROM employees
    GROUP BY department
    """,
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT MAX(salary) AS highest_salary FROM employees",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT MIN(salary) AS lowest_salary FROM employees",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) AS total_employees FROM employees",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT SUM(salary) AS total_salary FROM employees",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees WHERE name LIKE 'A%'",
    conn
)

df
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees WHERE salary BETWEEN 50000 AND 70000",
    conn
)

df
```

```python
cursor.execute(
    "UPDATE employees SET salary = 75000 WHERE name = 'John'"
)

conn.commit()

print("Salary updated successfully!")
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees",
    conn
)

df
```

```python
cursor.execute(
    "DELETE FROM employees WHERE name = 'Bob'"
)

conn.commit()

print("Record deleted successfully!")
```

```python
df = pd.read_sql_query(
    "SELECT * FROM employees",
    conn
)

df
```

```python
conn.close()

print("Database connection closed successfully!")
```
