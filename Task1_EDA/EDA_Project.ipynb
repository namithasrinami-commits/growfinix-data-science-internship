```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=30)
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
sns.boxplot(x='bedrooms', y='price', data=df)
plt.show()
```

# Real Estate Exploratory Data Analysis

## Objective
To analyze housing price data and understand relationships between area, bedrooms, bathrooms, stories, and price.

## Libraries Used
- Pandas
- NumPy
- Matplotlib
- Seaborn

## Analysis Performed
- Dataset information
- Missing value check
- Statistical summary
- Price distribution analysis
- Correlation heatmap
- Bedrooms vs Price analysis

## Conclusion
Larger houses with more bedrooms and bathrooms generally have higher prices. Area and number of stories also influence house prices.
