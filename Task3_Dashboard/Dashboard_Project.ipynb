```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales": [15000,18000,22000,25000,28000,32000]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o")
plt.title("Monthly Sales Dashboard")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Month", y="Sales", data=df)
plt.title("Monthly Sales Bar Chart")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df["Sales"], labels=df["Month"], autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.show()
```

# Sales Dashboard Analysis

## Objective
To visualize monthly sales data using different charts.

## Charts Used
- Line Chart
- Bar Chart
- Pie Chart

## Conclusion
Sales show a steady increase from January to June. June has the highest sales, indicating positive business growth.
